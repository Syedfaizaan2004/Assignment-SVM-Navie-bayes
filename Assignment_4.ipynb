# Pratical Questions 

**Q(1) -- What is a Support Vector Machine SVM?**

**Answer:** SVM is a supervised learning algorithm used for classification and regression.
It finds the optimal hyperplane that best separates data points of different classes with the maximum margin.

**Q(2) -- What is the difference between Hard Margin and Soft Margin SVM**

**Answer** 

**Hard Margin SVM**

Assumes data is perfectly separable; no misclassification allowed. May overfit or fail if data has noise.

**Soft Margin SVM**

Allows some misclassification using a penalty term more flexible and robust to noise.

**Q(3) -- What is the mathematical intuition behind SVM?**

**Answer** The mathematical instuition behind SVM is the goal is to maximize the margin between classes.

**Q(4) -- What is the role of Lagrange Multipliers in SVM?**

**Answer** Lagrange multipliers are used to convert the constrained optimization problem into an unconstrained dual problem. This allows solving SVM efficiently and helps identify support vectors where the constraints are active.

**Q(5) --  What are Support Vectors in SVM?**

**Answer:** Support vectors are the data points closest to the hyperplane. They are critical in defining the decision boundary; removing them would change the model.

**Q(6) -- What is a Support Vector Classifier (SVC)?**

**Answer:** SVC is the classification version of SVM. It finds the optimal hyperplane that separates classes with the maximum margin.

**Q(7) -- What is a Support Vector Regressor (SVR)?**

**Answer** SVR is the regression version of SVM. It fits a line (or hyperplane) such that errors within a certain margin (ε) are ignored, focusing on points outside the margin.

**Q(8) -- What is the Kernel Trick in SVM?**

**Answer** The kernel trick allows SVM to handle non-linearly separable data by mapping input features into a higher-dimensional space without explicitly computing the transformation. Common kernels: Linear, Polynomial, RBF, Sigmoid.

**Q(9) -- Compare Linear Kernel, Polynomial Kernel, and RBF Kernel?**

**Answer** Linear:
- No transformation; works well for linearly separable data.
- Text or high-dimensional data.

Polynomial:
- Adds interaction features up to a given degree.
- When features have polynomial relationships.

RBF (Gaussian):
- Measures similarity using distance; handles complex nonlinear boundaries.
- Most common for general nonlinear problems.

**Q(10) -- What is the effect of the C parameter in SVM?**

**Answer** Effect of the C Parameter in SVM: 
- Controls the trade-off between margin width and classification error.
- High C → less tolerance for misclassification (smaller margin, risk of overfitting).
- Low C → more tolerance (wider margin, risk of underfitting).

**Q(11) -- What is the role of the Gamma parameter in RBF Kernel SVM?**

**Answer**  Role of Gamma in RBF Kernel SVM:
- Gamma defines how far the influence of a single training example reaches.
- High Gamma → close influence, more complex boundary (overfitting).
- Low Gamma → broader influence, smoother boundary (underfitting).

**Q(12) -- What is the Naïve Bayes classifier, and why is it called "Naïve"?**

**Answer** Naïve Bayes is a probabilistic classifier based on Bayes’ Theorem, assuming features are independent given the class label.
It’s called “Naïve” because in reality, this independence rarely holds — but it still performs well.

**Q(13) -- What is Bayes’ Theorem?**

**Answer**   P(A|B) = (P(B|A) * P(A))/P(B)
- It calculates the posterior probability P(A|B) using prior probabilty P(A). likelihood P(B|A), and evidence P(B).

**Q(14) -- Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes?**

**Answer.**  
| Type               | Suitable For    | Description                                                 |
| ------------------ | --------------- | ----------------------------------------------------------- |
| **Gaussian NB**    | Continuous data | Assumes features follow a normal distribution.              |
| **Multinomial NB** | Count data      | Used in text classification (word counts, TF-IDF).          |
| **Bernoulli NB**   | Binary data     | Used for features with 0/1 presence (e.g., spam detection). |


**Q(15) -- When should you use Gaussian Naïve Bayes over other variants?**

**Answer** Use when features are continuous and approximately normally distributed — e.g., height, weight, or sensor readings.

**Q(16) -- What are the key assumptions made by Naïve Bayes?**

**Answer**  The key assumptions made by Naive Bayes are:
1. All features are conditionally independent given the class.
2. Each feature contributes equally to the outcome.
3. Dataset features are correctly represented by the chosen distribution.

**Q(17) -- What are the advantages and disadvantages of Naïve Bayes?**

**Answer**   Advantages:
1. Simple, fast and easy to implement.
2. Works well on small datasets.
3. Effective for high-dimensional data.

Disadvantages:
1. Assumes feature independence.
2. Performs body if features are highly correlated.

**Q(18) -- Why is Naïve Bayes a good choice for text classification?**

**Answer:** The Naive Bayes a good choice for text classification:
- Text data is high-dimensional and sparse, where feature independence is roughly valid.
- Efficient and fast for large vocabularies.
- Works well with word frequency features.

**Q(19) -- Compare SVM and Naïve Bayes for classification tasks?**

**Answer**  
| **Aspect**                | **SVM**                     | **Naïve Bayes**                       |
| --------------------- | --------------------------- | ------------------------------------- |
| **Type**              | Discriminative              | Generative                            |
| **Decision Boundary** | Finds optimal hyperplane    | Uses probability model                |
| **Performance**       | Good for complex boundaries | Good for text / high-dimensional data |
| **Speed**             | Slower to train             | Very fast                             |
| **Interpretability**  | Moderate                    | High (probabilities)                  |


**Q(20) -- How does Laplace Smoothing help in Naïve Bayes?**

**Answer**  Laplace smoothing avoids zero probabilty issues when a feature doesn't appear in the training data for a class. it adjusts counts as: 

P(word|class) = (count(word, class) + 1) / ((total_words_in_class) + V)
- where V = number of unique words.

Q(21) -- Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy?

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
svm_clf = SVC(kernel='linear', random_state=42)
svm_clf.fit(X_train, y_train)
y_pred = svm_clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

Q(24) -- Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary?

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.svm import SVC

X, y = make_moons(noise=0.2, random_state=42)
model = SVC(kernel='poly', degree=3)
model.fit(X, y)

import numpy as np
xx, yy = np.meshgrid(np.linspace(-2, 3, 300), np.linspace(-1.5, 2, 300))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
plt.title("SVM with Polynomial Kernel")
plt.show()
```

Q(27) Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually?

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.svm import SVC
import numpy as np

X, y = make_classification(n_features=2, n_redundant=0, n_informative=2, random_state=42)
C_values = [0.1, 1, 10]

for i, C in enumerate(C_values, 1):
    model = SVC(kernel='linear', C=C).fit(X, y)
    xx, yy = np.meshgrid(np.linspace(-3, 3, 200), np.linspace(-3, 3, 200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    plt.subplot(1, 3, i)
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
    plt.title(f"C={C}")
plt.show()
```

Q(29) Write a Python program to apply feature scaling before training an SVM model and compare results with unscaled data?

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Without scaling
model1 = SVC().fit(X_train, y_train)
acc1 = accuracy_score(y_test, model1.predict(X_test))

# With scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
model2 = SVC().fit(X_train_scaled, y_train)
acc2 = accuracy_score(y_test, model2.predict(X_test_scaled))

print("Without Scaling:", acc1)
print("With Scaling:", acc2)
```

Q(30) Write a Python program to train a Gaussian Naïve Bayes model and compare the predictions before and after Laplace Smoothing?

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

# Normal Gaussian NB
model1 = GaussianNB()
model1.fit(X_train, y_train)
acc1 = accuracy_score(y_test, model1.predict(X_test))

# Gaussian NB with Laplace smoothing (var_smoothing parameter)
model2 = GaussianNB(var_smoothing=1e-5)
model2.fit(X_train, y_train)
acc2 = accuracy_score(y_test, model2.predict(X_test))

print("Without Laplace Smoothing:", acc1)
print("With Laplace Smoothing:", acc2)
```

Q(31) Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C, gamma, kernel)? 

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

param_grid = {'C':[0.1,1,10], 'gamma':[0.001,0.01,0.1], 'kernel':['linear','rbf']}
grid = GridSearchCV(SVC(), param_grid, cv=5)
grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best Accuracy:", accuracy_score(y_test, grid.predict(X_test)))
```

Q(32) Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and check it improve accuracy?

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = make_classification(n_classes=2, weights=[0.9, 0.1], n_samples=1000, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Without class weights
model1 = SVC()
model1.fit(X_train, y_train)
acc1 = accuracy_score(y_test, model1.predict(X_test))

# With balanced class weights
model2 = SVC(class_weight='balanced')
model2.fit(X_train, y_train)
acc2 = accuracy_score(y_test, model2.predict(X_test))

print("Without Class Weighting:", acc1)
print("With Class Weighting:", acc2)
```

Q(38)  Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the average accuracy?

Q(39)  Write a Python program to train a Naïve Bayes classifier using different prior probabilities and compare
performance?

```python
from sklearn.datasets import load_breast_cancer
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

priors = [[0.3, 0.7], [0.5, 0.5]]
for p in priors:
    model = GaussianNB(priors=p)
    model.fit(X_train, y_train)
    print(f"Priors {p} → Accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

Q(40)  Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and compare accuracy?

```python
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score

X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

rfe = RFE(SVC(kernel='linear'), n_features_to_select=10)
X_train_rfe = rfe.fit_transform(X_train, y_train)
X_test_rfe = rfe.transform(X_test)

model = SVC(kernel='linear')
model.fit(X_train_rfe, y_train)
print("Accuracy after RFE:", accuracy_score(y_test, model.predict(X_test_rfe)))
```

Q(46)  Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve?

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

model = SVC(probability=True)
model.fit(X_train, y_train)
y_prob = model.predict_proba(X_test)[:, 1]

precision, recall, _ = precision_recall_curve(y_test, y_prob)
plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (SVM)")
plt.show()
```
